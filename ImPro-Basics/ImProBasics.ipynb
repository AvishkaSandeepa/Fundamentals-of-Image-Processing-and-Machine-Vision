# Basic Array of pixels
## Gray Scale pixel array creation

```python
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

im = np.zeros((6,8), dtype = np.uint8) 
im[2,3] = 150

print(im[2,3])

fig, axes = plt.subplots()
axes.imshow(im, cmap = "gray", vmin = 0, vmax = 255) #make sure to set pixel limits
axes.set_title("Gray Scale Image")
axes.xaxis.tick_top()

plt.show()
```

## Color pixel array creation

```python
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

im = np.zeros((6,8,3), dtype = np.uint8) 
im[2,3] = [255,0,0]  # first value for RED, second value for GREEN, third value for BLUE

print(im[2,3])

fig, axes = plt.subplots()
axes.imshow(im, cmap = "gray", vmin = 0, vmax = 255) #make sure to set pixel limits
axes.set_title("Color Image")
axes.xaxis.tick_top()

plt.show()
```

# Displaying an Images

```python
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img_original = cv.imread("./images/ema1.jpg", cv.IMREAD_COLOR) # dot is used to define location one step previous than current ipynb file
img_convert = cv.cvtColor(img_original, cv.COLOR_RGB2BGR)

fig, axes = plt.subplots(1,2, figsize = (20,20))
axes[0].imshow(img_original)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(img_convert)
axes[1].set_title("Converted Image")
axes[1].axis("off")
```

# Displaying with openCV's HighGUI

```python
import cv2 as cv
import matplotlib.pyplot as plt

img_original = cv.imread("./images/ema1.jpg", cv.IMREAD_COLOR)
cv.namedWindow("IMAGE", cv.WINDOW_AUTOSIZE) # create a window for displaying images.
cv.imshow("IMAGE", img_original) # show image in created window.....no need of colour conversion
cv.waitKey(0)
cv.destroyAllWindows()
```
