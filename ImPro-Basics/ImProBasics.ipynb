```python
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

im = np.zeros((6,8), dtype = np.uint8)
im[2,3] = 155

print(im[2,3])

fig, axes = plt.subplots()
axes.imshow(im, cmap = "gray", vmin = 0, vmax = 255)
axes.set_title("Gray Scale Image")
axes.xaxis.tick_top()

plt.show()
```
